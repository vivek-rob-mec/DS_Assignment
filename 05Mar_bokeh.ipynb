```python
# Q1. How can you create a Bokeh plot using Python code?
import bokeh.io
import bokeh.plotting
from bokeh.plotting import output_notebook, figure, show
output_notebook() # output file ploty.html
p = figure(title="Basic Line Plot", x_axis_label='x', y_axis_label='y')

p.line([1, 2, 3, 4], [4, 7, 2, 5], line_width=2)

show(p)
```

```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

# Glyphs are basic visual shapes (like circles, lines, rectangles) used to represent data points in Bokeh plots.

p = figure(title="cirle glyph example")
p.circle([1, 2, 3, 4], [4, 7, 2, 5], size=20, color="navy", alpha=0.5) # added circle as shape
show(p)
```

```python
# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
p = figure(title="Basic Line Plot", x_axis_label='x', y_axis_label='y')

p.line([1, 2, 3, 4], [4, 7, 2, 5], legend_label="Line A", line_color="green", line_width=3)
p.title.text_font_size = '16pt'
p.xaxis.axis_label_text_font_size = '14pt'
p.yaxis.axis_label_text_font_size = '14pt'
p.legend.location = 'top_left'
show(p)
```

```python
# Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
# real time?

# A Bokeh Server lets you build interactive, real-time visualizations that respond to user input or live data.

from bokeh.layouts import column
from bokeh.models import Slider
from bokeh.plotting import figure, curdoc

# Create a plot
plot = figure()
line = plot.line([1, 2, 3], [1, 4, 9])

slider = Slider(start=0.1, end=10, value=1, step=0.1, title="Scale")

def update(attr, old, new):
    scale = slider.value
    line.data_source.data['y'] = [i**2 * scale for i in [1, 2, 3]]

slider.on_change('value', update)

curdoc().add_root(column(slider, plot))
```

```python
# Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?
# app.py
from flask import Flask, render_template
from bokeh.embed import components
from bokeh.plotting import figure

app = Flask(__name__)

@app.route('/')
def home():
    plot = figure()
    plot.line([1, 2, 3], [4, 6, 5])
    script, div = components(plot)
    return render_template("index.html", script=script, div=div)

# index.html (template)
'''
<html>
  <head>{{ script | safe }}</head>
  <body>{{ div | safe }}</body>
</html>
'''

if __name__ == '__main__':
    app.run(debug=True)
```
