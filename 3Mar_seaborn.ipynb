```python
# Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
"""
Seaborn library in Python is built on top of Matplotlib and is used for beautiful and informative statistical visualizations.

1) lineplot: sns.lineplot()
used for: display trends over continous data, like time series data

2) barplot: sns.barplot()
used for: compare mean values across different categories. Adds error bars automatically

3) heatmap(): sns.heatmap()
used for: visualize correlation matrices or any 2D matrix-like data with color gradients. Excellent for spotting patterns and relationships.

4) boxplot(): sns.boxplot()
used for: to show the distribution, median, quartiles, and outliers of a dataset. Great for comparing distributions between groups.

5) scatterplot(): sns.scatterplot()
used for: display the relationship between two continuous variables. Can include color/hue for additional category-based separation.
"""
```

```python
# Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using
# x ="timepoint" and y = "signal" for different events and regions.
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from seaborn import load_dataset

df = load_dataset('fmri')
df.head()
sns.lineplot(x='timepoint',y='signal', data=df, hue='event', style='region')
# hue: to separate event using different color
# style: line style
plt.show()
```

```python
# Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
# 'pclass', y = 'age' and y = 'fare'.

df = load_dataset('titanic')
df.head()
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.boxplot(x='pclass', y='age', data=df)
plt.subplot(1,2,2)
sns.boxplot(x='pclass', y='fare', data=df)
plt.show()
```

```python
# Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
# parameter for the 'cut' column of the diamonds dataset.
df = load_dataset('diamonds')
df.head()
sns.histplot(data=df, x='price', hue='cut')
plt.show()
```

```python
# Q5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
# of the iris dataset.
df = load_dataset('iris')
df.head()
sns.pairplot(df, hue='species')
plt.show()
```

```python
# Q6: Use the "flights" dataset from seaborn to plot a heatmap.
df = load_dataset('flights')
import pandas as pd
df.head()
flights = df.pivot(index='month',columns='year',values='passengers')
sns.heatmap(flights, annot=True, fmt='d')
plt.show()
```
