```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot')
plt.show()
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
"""
subplot(): The subplot() function in Matplotlib is used to create multiple plots in a single figure.
           It helps in organizing multiple visualizations side by side or in a grid format.

           The subplot() function takes three arguments:
           1) nrows: The number of rows in the subplot grid.
           2) ncols: The number of columns in the subplot grid.
           3) index: The index of the subplot to be created.

           plt.subplot(nrows, ncols, index)

"""
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
# import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
import matplotlib.pyplot as plt
plt.bar(company,profit)
plt.show()
plt.barh(company,profit)
plt.show()

"""
barplot: displays categorical data with rectangular bars. The height (or length) of each bar represents the value or frequency of that category.

used for:
1) compare values for different categories
2) visualize frequency or count
3) identify trends pattern outliers
"""
```

```python
#Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
plt.show()
"""
boxplot(): boxplot() is a function in Matplotlib used to create a box-and-whisker plot. It provides a summary of the distribution of a dataset by showing:
1) Minimum
2) First quartile (Q1)
3) Median (Q2)
4) Third quartile (Q3)
5) Maximum
6) Outliers (if any)

used for:
1) to visualize spread and central tendency of data
2) to detect outliers
3) to compare distribution across multiple groups
"""
```
