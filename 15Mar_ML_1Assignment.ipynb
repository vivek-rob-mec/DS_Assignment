### Q2: What is supervised learning? List some examples of supervised learning.
Ans: \
**Definition:**  
Supervised learning is a type of **Machine Learning** where the model is trained on a **labeled dataset** — meaning each training example is paired with the correct output (label).  
The goal is for the model to learn a mapping from inputs to outputs, so it can predict the label for new, unseen data.

---

### **Key Points:**
- Input + Output (Label) → Train the model  
- The model “learns” from known data  
- Used for **classification** and **regression** tasks  

---

### **Examples of Supervised Learning:**

| **Example** | **Input** | **Output (Label)** | **Type** |
|-------------|-----------|--------------------|----------|
| **Email Spam Detection** | Email content | Spam / Not Spam | Classification |
| **House Price Prediction** | Size, location, age of house | Price of the house | Regression |
| **Handwriting Recognition** | Image of a digit | Number (0–9) | Classification |
| **Loan Approval** | Income, credit score, loan amount | Approve / Deny | Classification |
| **Weather Forecasting** | Temperature, humidity, pressure | Tomorrow’s temperature | Regression |

### Q3: What is unsupervised learning? List some examples of unsupervised learning.
Ans: \
**Definition:**  
Unsupervised learning is a type of **Machine Learning** where the model is trained on **unlabeled data** — meaning the data has **no predefined output**.  
The goal is to find hidden patterns, structures, or relationships in the data.

---

### **Key Points:**
- **No labels** provided  
- The model explores the data and **groups** or **organizes** it  
- Common tasks: **Clustering** and **Dimensionality Reduction**

---

### **Examples of Unsupervised Learning:**

| **Example** | **Input Data** | **Goal / Output** | **Type** |
|-------------|----------------|-------------------|----------|
| **Customer Segmentation** | Customer purchase history | Group similar customers | Clustering |
| **Market Basket Analysis** | Transaction data | Find items frequently bought together | Association |
| **Anomaly Detection** | Server logs | Identify unusual patterns (e.g., fraud) | Clustering / Outlier Detection |
| **Document Topic Modeling** | Collection of articles | Discover topics in texts | Clustering |
| **Image Compression** | Raw image data | Reduce image size without much quality loss | Dimensionality Reduction |

### Q5: What are the main differences between supervised, unsupervised, and semi-supervised learning?
Ans: \
### **1. Supervised Learning**

- **What it is:** The model learns from a **labeled dataset** — each input has a correct output.
- **Goal:** Predict the correct label for new data.
- **Used for:** Classification (e.g., spam detection) and regression (e.g., price prediction).

>  **Example:** You train a model with images of cats and dogs, and each image is labeled as either "cat" or "dog."

---

### **2. Unsupervised Learning**

- **What it is:** The model is given **unlabeled data**, and it tries to find hidden patterns or structures.
- **Goal:** Group or organize the data in meaningful ways.
- **Used for:** Clustering, dimensionality reduction, anomaly detection.

>  **Example:** You give a model a bunch of customer data (without labels), and it groups customers based on purchasing behavior.

---

### **3. Semi-Supervised Learning**

- **What it is:** A mix of **a small amount of labeled data** and **a large amount of unlabeled data**.
- **Goal:** Use the small labeled portion to help learn patterns in the unlabeled data.
- **Used for:** When labeling is expensive or time-consuming, but lots of unlabeled data is available.

>  **Example:** You have 1,000 images where only 100 are labeled as "cat" or "dog." The model learns from both the labeled and unlabeled images.

---

###  Key Differences:

- **Supervised** = Labeled data  
- **Unsupervised** = Unlabeled data  
- **Semi-Supervised** = Mostly unlabeled data + small labeled portion

### Q6: What is train, test and validation split? Explain the importance of each term.
Ans: \
When building a machine learning model, we split the dataset into **three parts** to ensure the model learns well, performs accurately, and generalizes to new data. These parts are:

---

### **1. Training Set**

- **What it is:**  
  The data the model actually learns from. It includes both inputs and their correct outputs (labels, in supervised learning).

- **Importance:**  
  This is where the model finds patterns, relationships, and adjusts its internal settings (like weights in neural networks).

>  Think of it as a student learning from a textbook and practice problems.

---

### **2. Validation Set**

- **What it is:**  
  A separate part of the data used during training to tune model parameters and settings (like learning rate, number of layers, etc.).

- **Importance:**  
  Helps prevent **overfitting** — when the model learns the training data too well but performs poorly on new data.  
  It guides model improvement during the training phase without using test data.

>  Like taking practice tests to see how well you're doing before the final exam.

---

### **3. Test Set**

- **What it is:**  
  This is **completely unseen data** — not used during training or tuning. It is used only **after** training is complete.

- **Importance:**  
  It gives a final, honest check of how well the model will perform in the real world. It shows the model’s ability to generalize.

>  Like the final exam — you haven’t seen the questions before, and it truly measures your understanding.

---

###  Why Splitting is Important:

- **Avoids overfitting** and gives better generalization.
- Ensures that model evaluation is fair and unbiased.
- Helps tune and compare different models or techniques safely.

### Q7: How can unsupervised learning be used in anomaly detection?
Ans: \

**Anomaly detection** means identifying data points that are **significantly different** from the rest — these could be errors, fraud, or rare events.

In **unsupervised learning**, you **don’t have labels**, so the model doesn’t know in advance which data is "normal" or "anomalous." Instead, it learns the structure or distribution of the data and flags anything that doesn’t fit the pattern.

---

### **How It Works:**

1. **Model learns normal patterns** from the data using techniques like:
   - **Clustering** (e.g., K-Means)
   - **Dimensionality reduction** (e.g., PCA)
   - **Autoencoders** (in deep learning)

2. **Anomalies** are identified as:
   - Data points that **don’t belong to any cluster**
   - Points that are **far from cluster centers**
   - Points with **high reconstruction error** (in case of autoencoders)

---

### **Example Use Cases:**

- **Banking:** Detecting fraudulent transactions that don’t match normal spending behavior.
- **Cybersecurity:** Spotting unusual login activity that might indicate a breach.
- **Healthcare:** Identifying abnormal patient vital signs.
- **Manufacturing:** Detecting equipment failures or defects by unusual sensor readings.

---

###  Why Unsupervised Learning?

- Anomalies are often **rare and unpredictable**, so labeled examples are hard to find.
- Unsupervised methods **don’t need labels**, making them perfect when you only have raw data.

### Q8: List down some commonly used supervised learning algorithms and unsupervised learning algorithms.
Ans: \

###  **Supervised Learning Algorithms:**
These require **labeled data** (input-output pairs).

1. **Linear Regression**  
   - Used for predicting continuous values (e.g., house price prediction)

2. **Logistic Regression**  
   - Used for binary classification (e.g., spam or not spam)

3. **Decision Trees**  
   - Easy-to-understand models for both classification and regression

4. **Random Forest**  
   - Ensemble of decision trees, improves accuracy

5. **Support Vector Machines (SVM)**  
   - Great for classification, especially with clear margin separation

6. **K-Nearest Neighbors (KNN)**  
   - Classifies based on the labels of nearest neighbors

7. **Naive Bayes**  
   - Probabilistic classifier based on Bayes’ theorem

8. **Gradient Boosting (e.g., XGBoost, LightGBM)**  
   - Powerful and commonly used for structured data

---

###  **Unsupervised Learning Algorithms:**
These work with **unlabeled data**.

1. **K-Means Clustering**  
   - Groups similar data points into *k* clusters

2. **Hierarchical Clustering**  
   - Builds a hierarchy/tree of clusters

3. **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**  
   - Finds clusters of varying shapes and identifies noise (anomalies)

4. **Principal Component Analysis (PCA)**  
   - Reduces the dimensionality of data, helps with visualization or preprocessing

5. **Autoencoders**  
   - Neural networks that learn to compress and reconstruct data (used for anomaly detection, denoising)

6. **t-SNE (t-distributed Stochastic Neighbor Embedding)**  
   - Used for visualizing high-dimensional data in 2D/3D

7. **Apriori / FP-Growth**  
   - Used for market basket analysis (association rule mining)
