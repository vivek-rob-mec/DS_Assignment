### Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in predicting the quality of wine.
Ans: \
Below are the main features (independent variables) commonly found in the dataset:

| Feature Name | Description | Importance |
|--------------|-------------|------------|
| **fixed acidity** | Tartaric acid content | Affects taste and pH; contributes to wine’s stability and freshness |
| **volatile acidity** | Acetic acid content | Too much gives a vinegar taste; negatively affects quality |
| **citric acid** | Adds freshness and flavor | Moderate levels enhance flavor, high levels might affect taste |
| **residual sugar** | Leftover sugar after fermentation | Contributes to sweetness; balance is key |
| **chlorides** | Salt content | Excessive saltiness can reduce wine quality |
| **free sulfur dioxide** | SO₂ not bound to other molecules | Acts as an antioxidant and antimicrobial agent |
| **total sulfur dioxide** | Free + bound SO₂ | Excess can negatively affect taste and smell |
| **density** | Related to sugar and alcohol content | Can indicate sugar/alcohol levels; indirectly affects quality |
| **pH** | Acidity level | Influences microbial stability and flavor profile |
| **sulphates** | Additives for preservation | Helps preserve and stabilize the wine; too much affects flavor |
| **alcohol** | Ethanol content | One of the strongest influencers of perceived quality |
| **quality** | Wine quality (score between 0-10) | Target variable to predict (subjective human rating) |

### Importance of Each Feature in Predicting Wine Quality

1. **Alcohol** – Often shows strong positive correlation with quality (more alcohol → higher perceived quality).
2. **Volatile Acidity** – Usually has a strong **negative** correlation with quality (high acidity lowers quality).
3. **Sulphates** – Acts as a preservative and may improve stability and taste when balanced.
4. **Density, pH** – Affect fermentation and storage quality.
5. **Residual Sugar** – Important for taste; both too much and too little can impact quality negatively.
6. **Citric & Fixed Acidity** – Influence freshness and tartness of wine, contributing to flavor.


### Q2. How did you handle missing data in the wine quality data set during the feature engineering process? Discuss the advantages and disadvantages of different imputation techniques.
Ans: \

###  **Step-by-Step: Handling Missing Data**

1. **Identify Missing Values**  
   Use functions like `df.isnull().sum()` or `df.info()` in pandas to locate missing data.

2. **Assess the Proportion of Missing Data**  
   - **Low (<5%)**: Likely safe to drop or impute.
   - **Medium (5–30%)**: Prefer imputation.
   - **High (>30%)**: Consider dropping the feature or advanced imputation.

---

###  **Common Imputation Techniques**

#### 1. **Mean/Median Imputation**
- Replace missing values with the mean or median of the column.

**Advantages:**
- Simple and fast.
- Preserves dataset size.

**Disadvantages:**
- Can distort data distribution.
- Doesn’t handle outliers well (mean is sensitive).

#### 2. **Mode Imputation** (for categorical features)
- Use the most frequent value.

**Good for** features like wine type (red/white) or country of origin.

#### 3. **K-Nearest Neighbors (KNN) Imputation**
- Finds similar rows using distance metrics and fills missing values based on neighbors.

**Advantages:**
- More accurate for small-to-medium datasets.
- Preserves patterns and variability.

**Disadvantages:**
- Computationally expensive.
- Can be sensitive to outliers or irrelevant features.

#### 4. **Multivariate Imputation (e.g., MICE)**
- Models each feature with missing data as a function of other features.

**Advantages:**
- Statistically powerful.
- Suitable for datasets with multiple correlated variables.

**Disadvantages:**
- Complex and time-consuming.
- Needs careful tuning.

#### 5. **Drop Missing Rows or Columns**
- Safe when the missingness is random and sparse.

**Advantages:**
- Simplifies data pipeline.

**Disadvantages:**
- Loses potentially useful data.
- Risk of bias if missingness is not random.

---

###  Example in Python (Mean Imputation)
```python
from sklearn.impute import SimpleImputer
import pandas as pd

# Assuming df is the wine dataset with missing values
imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
```

### Q4. Describe the process of feature engineering in the context of the student performance data set. How did you select and transform the variables for your model?

* Feature engineering is the process of selecting, creating, transforming, or encoding variables (features) to improve the performance of a machine learning model.


###  **Feature Engineering Process in Student Performance Dataset**

---

###  Step 1: **Understand the Dataset**

Common features in a student performance dataset may include:

- Demographic: `gender`, `age`, `address (urban/rural)`
- Academic: `studytime`, `failures`, `absences`, `schoolsup` (extra support), `G1`, `G2`, `G3` (grades)
- Social: `internet access`, `freetime`, `health`, `famrel` (family relationship)

---

###  Step 2: **Feature Selection**

**Objective**: Predict final exam score (`G3`) or classify students as pass/fail.

#### ✔ Features to include:

- **Past performance**: `G1`, `G2` (strong predictors of `G3`)
- **Support systems**: `schoolsup`, `famsup`, `paid`

#### ✘ Features to possibly exclude:
- **Redundant or weakly correlated features** (e.g., `school`, `Mjob`, `Fjob` if correlation is low)
- Features with **high cardinality but low information value**

---

###  Step 3: **Feature Transformation**

####  1. **Encoding Categorical Variables**
- **Binary Encoding**:
  - `gender`: M → 0, F → 1
  - `schoolsup`: yes → 1, no → 0

- **One-Hot Encoding** for multi-class:
  - `guardian`: mother, father, other
  - `internet`: yes/no → binary

```python
df['gender'] = df['gender'].map({'M': 0, 'F': 1})
df = pd.get_dummies(df, columns=['guardian'], drop_first=True)
```

---

####  2. **Handling Missing Data**
- If present, impute using **mean/median** for numerical features or **mode** for categorical.

---

####  3. **Creating New Features**
- `average_grade = (G1 + G2 + G3) / 3`
- `improved = 1 if G3 > G1 else 0` (did student improve?)

####  4. **Feature Scaling**
- Normalize `studytime`, `absences`, `G1`, `G2`, `G3` using **Min-Max scaling** or **StandardScaler**.

```python
from sklearn.preprocessing import MinMaxScaler

df[['studytime', 'absences']] = scaler.fit_transform(df[['studytime', 'absences']])
```

---

####  5. **Outlier Detection & Treatment**
- Remove or cap extreme values in features like `absences`, `G3`

---

###  Step 4: **Feature Importance**
Use models (like Random Forest or XGBoost) to rank features and keep the most important ones.

```python
from sklearn.ensemble import RandomForestRegressor

model.fit(X_train, y_train)
```

---

## Summary

- Carefully **select relevant variables** based on domain knowledge and correlation.
- **Transform categorical and numerical variables** appropriately.
- Create new features that **capture deeper patterns** (like grade improvement or support).
- Always **scale features** and **remove noise/outliers**.

###  **EDA and Normality Check for Wine Quality Dataset**

---

###  Step 1: **Load the Dataset**

The dataset is usually available as `winequality-red.csv` or `winequality-white.csv`.

```python
import pandas as pd
```

###  Step 2: **Check Feature Distribution**

Use **histograms** or **distplots** to visually assess the distribution.

```python
for col in df.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

You can also compute **skewness**:

```python
df.skew().sort_values(ascending=False)
```

> **Interpretation**:
- Skewness > +1 or < -1 → Highly skewed (non-normal)
- Between ±0.5 → Approximately normal

---

###  Step 3: **Identify Non-Normal Features**

Typical features that exhibit non-normality in wine datasets include:

| Feature               | Typical Skew | Comment |
|----------------------|--------------|---------|
| `residual sugar`     | Right (positive skew) | Long tail due to sweet wines |
| `chlorides`          | Right skew   | Very small values with few high outliers |
| `free sulfur dioxide`| Right skew   | High variance |
| `total sulfur dioxide` | Right skew | Can vary widely |
| `alcohol`            | Slight skew  | Sometimes closer to normal |

---

###  Step 4: **Transform Skewed Features**

To improve normality, apply transformations:

#### ✔ Log Transformation (for right-skewed):
```python
import numpy as np

df['residual sugar'] = np.log1p(df['residual sugar'])
df['chlorides'] = np.log1p(df['chlorides'])
df['total sulfur dioxide'] = np.log1p(df['total sulfur dioxide'])
```

#### ✔ Box-Cox Transformation (if no zero/negative values):
```python
from scipy.stats import boxcox

df['chlorides'], _ = boxcox(df['chlorides'] + 1e-6)  # add small value if zero present
```

#### ✔ Yeo-Johnson (can handle zero/negative):
```python
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method='yeo-johnson')
df_transformed = pd.DataFrame(pt.fit_transform(df), columns=df.columns)
```

---

###  Step 5: **Re-check Distributions**

Plot the histograms again after transformation to verify the changes.

---

###  Summary

- **Non-normal features** often include `residual sugar`, `chlorides`, `sulfur dioxide levels`.
- Use **skewness values** and **visual plots** to identify them.
- **Log, Box-Cox, or Yeo-Johnson** transformations can help improve normality.
- Normalizing features helps many ML models perform better, especially those that assume Gaussian distributions (e.g., linear regression, logistic regression, LDA).

### Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?
Ans: \


````python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/winequality-red.csv')  # or winequality-white.csv

# Separate features and target
X = df.drop('quality', axis=1)

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Explained variance
explained_var = pca.explained_variance_ratio_

# Cumulative explained variance
cumulative_var = explained_var.cumsum()

# Plot
plt.figure(figsize=(8, 5))
sns.lineplot(x=range(1, len(cumulative_var)+1), y=cumulative_var, marker='o')
plt.axhline(y=0.9, color='r', linestyle='--')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance vs. Number of Components')
plt.grid(True)
plt.show()

import numpy as np

# Find how many components explain at least 90% variance
num_components_90 = np.argmax(cumulative_var >= 0.90) + 1
print(f"Number of components needed to explain 90% variance: {num_components_90}")
````
