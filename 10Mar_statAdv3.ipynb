# Q1: What is Estimation Statistics? Explain point estimate and interval estimate.
Ans: \
Estimation statistics is a branch of statistics that deals with inferring the value of a population parameter based on sample data. Rather than simply describing data (as in descriptive statistics), estimation aims to make predictions or generalizations about a larger group.

 Two Main Types of Estimation:
1. Point Estimate
* A single value used to estimate a population parameter.

* Example: The sample mean (ð‘¥Ì„) is a point estimate of the population mean (Î¼).

Example: If the average height of a sample of 100 students is 165 cm, then 165 cm is the point estimate of the average height of the entire student population.

2. Interval Estimate (Confidence Interval)
* An interval or range of values used to estimate a population parameter.

* It provides a confidence level, typically 95% or 99%, which reflects the probability that the interval contains the true parameter.

Example: A 95% confidence interval for the population mean might be (160 cm, 170 cm), meaning we are 95% confident the actual population mean lies within this range

````python
# Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.
import scipy.stats as stats
import math

def estimate_population_mean(sample_mean, sample_std, sample_size, confidence = 0.95):
    z_score = stats.norm.ppf((1 + confidence )/2)
    margin_error= z_score * (sample_std / math.sqrt(sample_size))

    lower_bound = sample_mean - margin_error
    upper_bound = sample_mean + margin_error

    return lower_bound, upper_bound

sample_mean = 50
sample_std = 10
sample_size = 42

lower_bound, upper_bound = estimate_population_mean(sample_mean, sample_std, sample_size)
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
````

# Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.
Ans: \

**Hypothesis Testing** is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves testing an assumption (called a *hypothesis*) about a population.

---

### Why is it used?

It is used to:
- Determine whether the observed data provides enough evidence to reject a stated assumption.
- Compare groups, variables, or treatments in an objective way.
- Assess the effectiveness of a new drug, marketing campaign, or teaching method, etc.

---

### Importance of Hypothesis Testing:

1. **Decision Making:** Helps in making data-driven decisions instead of relying on intuition.
2. **Validity Testing:** Validates or refutes claims about a population.
3. **Scientific Research:** Essential in experiments to prove or disprove theories.
4. **Risk Management:** Helps in identifying risks or differences before applying changes.
5. **Business Strategy:** Used in A/B testing, product comparison, and market research.

---

### Simple Example:

You want to test if a new teaching method improves studentsâ€™ test scores.  
- **Null Hypothesis (Hâ‚€):** The new method has no effect.
- **Alternative Hypothesis (Hâ‚):** The new method improves scores.  
By collecting sample data and performing a test (e.g., t-test), you can decide whether to reject Hâ‚€ or not.

#Q4. Create a hypothesis that states whether the average weight of male college students is greater than the average weight of female college students.
Ans: \
##Comparing Average Weights of Male and Female College Students

We want to test whether the **average weight of male college students is greater** than that of **female college students**.

---

###  Step-by-Step Hypotheses Setup:

Let:

- Î¼â‚ = average weight of **male** college students  
- Î¼â‚‚ = average weight of **female** college students

---

###  Hypotheses:

- **Null Hypothesis (Hâ‚€):** Î¼â‚ â‰¤ Î¼â‚‚  
  *(The average weight of male students is less than or equal to female students.)*

- **Alternative Hypothesis (Hâ‚):** Î¼â‚ > Î¼â‚‚  
  *(The average weight of male students is greater than female students.)*

---

This is a **one-tailed (right-tailed) test** because we are specifically testing if one mean is *greater than* the other.

You would collect weight data from random samples of both groups and apply a **two-sample t-test** to determine if there's statistically significant evidence to reject the null hypothesis.

````python
# Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
# given a sample from each population.
import numpy as np
from scipy import  stats
import math

sample1 = np.array([72, 75, 78, 70, 74, 77, 79])  # e.g., Male weights
sample2 = np.array([65, 67, 64, 63, 66, 62, 68])  # e.g., Female weights

male_mean = np.mean(sample1)
female_mean = np.mean(sample2)

male_std = np.std(sample1, ddof=1)
female_std = np.std(sample2, ddof=2)

print(f"male_mean: {male_mean: .2f}, female_mean: {female_mean: .2f}")
print(f"male_std: {male_std: .2f}, female_std: {female_std: .2f}")

t_stat, p_value = stats.ttest_ind(sample1,sample2, equal_var=False)

print(f"t_stat: {t_stat: .2f}, p_value: {p_value: .2f}")

alpha = 0.05
if p_value < alpha:
  print(f"Reject null hypothesis, since there is a significance difference betweeen two population mean")
else:
  print(f"Fail to reject null hypothesis, since there is no significance difference betweeen two population mean")

"""
Output Interpretation:
T-statistic tells how far the sample means are from each other relative to variability.

P-value tells the probability of observing such a difference assuming the null hypothesis is true.

If p < 0.05, we reject the null and conclude that the means are significantly different.
"""
````

# Q7: Write down the steps involved in hypothesis testing.
Ans: \
**standard steps involved in hypothesis testing**:

###  Step-by-Step Guide:

1. **State the Hypotheses**
   - **Null Hypothesis (Hâ‚€):** A statement of no effect or no difference.
   - **Alternative Hypothesis (Hâ‚ or Ha):** A statement that contradicts Hâ‚€.

2. **Set the Significance Level (Î±)**
   - Common values: `0.05`, `0.01`, or `0.10`
   - This is the probability of rejecting Hâ‚€ when it is actually true (Type I error).

3. **Choose the Appropriate Test**
   - Depends on the type of data and sample size:
     - z-test
     - t-test
     - chi-square test
     - ANOVA
     - etc.

4. **Calculate the Test Statistic**
   - Use formulas or Python libraries (e.g., `scipy.stats`)
   - Compare sample data to what is expected under Hâ‚€.

5. **Find the p-value**
   - The p-value tells you the probability of getting the observed result (or more extreme) if Hâ‚€ is true.

6. **Make a Decision**
   - If `p-value â‰¤ Î±`: **Reject Hâ‚€** (evidence supports Hâ‚)
   - If `p-value > Î±`: **Fail to reject Hâ‚€** (not enough evidence)

7. **State the Conclusion**
   - Interpret the results in the context of the original question.

---

### Example Summary:
> "At 5% significance level, since the p-value is less than 0.05, we reject the null hypothesis and conclude that the new drug significantly lowers blood pressure."

# Q8. Define p-value and explain its significance in hypothesis testing.
Ans:\
The p-value (probability value) is a number between 0 and 1 that helps you determine the strength of evidence against the null hypothesis (Hâ‚€) in hypothesis testing.

Definition:
The p-value is the probability of obtaining a test statistic at least as extreme as the one observed, assuming that the null hypothesis is true.


### Significance of the p-value:

| p-value | Interpretation |
|---------|----------------|
| â‰¤ Î± (e.g., 0.05) | **Strong evidence against Hâ‚€** â†’ Reject Hâ‚€ |
| > Î± | **Weak evidence against Hâ‚€** â†’ Fail to reject Hâ‚€ |

---

### Why is the p-value important?
- It quantifies **how likely** it is to get the observed result **by random chance**.
- It allows researchers to make **objective decisions** rather than guessing.
- Itâ€™s used in fields like medicine, engineering, social sciences, business analytics, etc.

---

### Example:
Suppose you're testing a new drug, and the p-value = **0.03**.
- If your significance level Î± = **0.05**, then:
  - Since `0.03 < 0.05`, you **reject the null hypothesis**.
  - Conclusion: The drug **has a statistically significant effect**.

# Q12: What is t-statistic? State the formula for t-statistic.
Ans: \
### **Q12: What is the t-statistic?**

The **t-statistic** (or **t-score**) is a value used in **hypothesis testing** to determine how far the sample mean is from the population mean, in terms of the **standard error**. It helps decide whether to reject the **null hypothesis**.

It is especially used when:
- The sample size is small.
- The population standard deviation is unknown.

---

### âœ… **Formula for t-statistic:**

For a **one-sample t-test**, the formula is:
$$
[
t = \frac{\bar{x} - \mu}{s / \sqrt{n}}
]
$$
Where:  
- \( bar{x} \) = sample mean
- \( mu \) = population mean (under the null hypothesis)  
- \( s \) = sample standard deviation  
- \( n \) = sample size  

---

###  **Use Case:**
If you want to test whether the average height of a group of 20 students differs from the national average, and you don't know the population standard deviation â€” you use the **t-statistic**.

# Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50. Estimate the population mean revenue with a 95% confidence interval.
Ans: \
To estimate the **population mean revenue** using a **95% confidence interval**, we can apply the formula for the confidence interval using the **t-distribution** (since the population standard deviation is unknown and the sample size is relatively small).

---

### **Given:**
- Sample size, \( n = 50 \)
- Sample mean, \( \bar{x} = 500 \)
- Sample standard deviation, \( s = 50 \)
- Confidence level = 95%
- Degrees of freedom = \( n - 1 = 49 \)

---

### **Formula for Confidence Interval:**

$$
[
\text{CI} = \bar{x} \pm t_{\alpha/2} \times \frac{s}{\sqrt{n}}
]
$$

Where:
- $$( t_{\alpha/2} )$$ is the critical t-value for 95% CI and 49 degrees of freedom
- $$ ( \frac{s}{\sqrt{n}} )$$ is the standard error

---

### **Python Code:**

```python
import scipy.stats as stats
import math

# Given values
sample_mean = 500
sample_std = 50
n = 50
confidence = 0.95

# Degrees of freedom
df = n - 1

# t critical value
t_critical = stats.t.ppf((1 + confidence) / 2, df)

# Standard error
se = sample_std / math.sqrt(n)

# Margin of error
margin_of_error = t_critical * se

# Confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

(lower_bound, upper_bound)
```

---

### **Interpretation:**
The 95% confidence interval will give you a range like:
[
(485.14, 514.86)
]

Which means: _"We are 95% confident that the true population mean daily revenue lies between \$485.14 and \$514.86."_

# Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.
Ans: \
To test the researcher's hypothesis, we can perform a **one-sample t-test** comparing the sample mean to the hypothesized mean.

---

### âœ… **Step-by-step Hypothesis Testing:**

#### **1. Define the hypotheses**

- **Null Hypothesis (Hâ‚€):**  
  The mean decrease in blood pressure is **10 mmHg**.  
  \( H_0: mu = 10 \)

- **Alternative Hypothesis (Hâ‚):**  
  The drug decreases blood pressure by **less than** 10 mmHg.  
  \( H_1: mu < 10 \) (Left-tailed test)

---

### âœ… **2. Given values:**

- Hypothesized mean \( mu_0 = 10 \)
- Sample mean \( bar{x} = 8 \)
- Sample standard deviation \( s = 3 \)
- Sample size \( n = 100 \)
- Significance level \( alpha = 0.05 \)

---

###  **3. Test statistic (t):**
$$
[
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}} = \frac{8 - 10}{3 / \sqrt{100}} = \frac{-2}{0.3} = -6.67
]
$$

---

###  **4. Critical t-value:**

Using \( df = n - 1 = 99 \), for a **left-tailed** test at \( alpha = 0.05 \):
$$
[
t_{\text{critical}} = -1.660 \quad \text{(approx)}
]
$$

---

### **5. Decision Rule:**
$$
 If ( t \leq t_{\text{critical}} ), reject ( H_0 )
$$
Since:
$$
[
t = -6.67 \leq -1.660
]
$$
* We **reject the null hypothesis**.

---

### **Conclusion:**

There is **statistically significant evidence** at the 0.05 level to conclude that the new drug **reduces blood pressure by less than 10 mmHg**.

### Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 = 30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two groups are equal with a significance level of 0.01.
Ans: \
We will perform a **two-sample t-test** (assuming unequal variances, also known as Welchâ€™s t-test) to test if the **population means are equal**.

---

### âœ… **Step-by-step Hypothesis Test**

#### **1. Define the Hypotheses**

- **Null Hypothesis (Hâ‚€):** \( mu_1 = mu_2 \) (The two groups have equal mean scores)
- **Alternative Hypothesis (Hâ‚):** \( mu_1 == mu_2 \) (The two groups have different mean scores)

This is a **two-tailed test**.

---

###  **2. Given:**

- \( bar{x}_1 = 80 \), \( s_1 = 10 \), \( n_1 = 30 \)
- \( bar{x}_2 = 75 \), \( s_2 = 8 \), \( n_2 = 40 \)
- Significance level \( alpha = 0.01 \)

---

### **3. Calculate the test statistic (Welchâ€™s t):**
$$
[
t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}
= \frac{80 - 75}{\sqrt{\frac{100}{30} + \frac{64}{40}}}
= \frac{5}{\sqrt{3.33 + 1.6}} = \frac{5}{\sqrt{4.93}} \approx \frac{5}{2.22} \approx 2.25
]
$$

---

###  **4. Degrees of Freedom (approx):**
$$
[
df \approx \frac{ \left( \frac{s_1^2}{n_1} + \frac{s_2^2}{n_2} \right)^2 }{ \frac{(\frac{s_1^2}{n_1})^2}{n_1 - 1} + \frac{(\frac{s_2^2}{n_2})^2}{n_2 - 1} }
= \frac{(4.93)^2}{\frac{(100/30)^2}{29} + \frac{(64/40)^2}{39}} \approx \frac{24.3}{0.372 + 0.105} \approx \frac{24.3}{0.477} \approx 51
]
$$

So, degrees of freedom â‰ˆ 51

---

### **5. Critical t-value:**

- Two-tailed test, \( alpha = 0.01 \)
- Degrees of freedom â‰ˆ 51

Using a t-distribution table or calculator:  
$$( t_{\text{critical}} \approx \pm 2.68 )$$

---

###  **6. Decision Rule:**

If $$ ( |t| > t_{\text{critical}} ), reject ( H_0 ) $$

Since:
$$
[
|t| = 2.25 < 2.68
]
$$

 We **fail to reject the null hypothesis**.

---

###  **Conclusion:**

At the 0.01 significance level, there is **not enough evidence** to conclude that the population means of the two groups are different.

### Q17. A marketing company wants to estimate the average number of ads watched by viewers during a TV program. They take a random sample of 50 viewers and find that the sample mean is 4 with a standard deviation of 1.5. Estimate the population mean with a 99% confidence interval.

To estimate the **population mean** with a **99% confidence interval**, we'll use the formula for the confidence interval when the population standard deviation is unknown and the sample size is small:

---

### âœ… **Given:**

- Sample size \( n = 50 \)  
- Sample mean \( bar{x} = 4 \)  
- Sample standard deviation \( s = 1.5 \)  
- Confidence level = 99%  
- Degrees of freedom \( df = n - 1 = 49 \)

---

###  **Step 1: Find the t-critical value**

At 99% confidence level and \( df = 49 \), the **t-critical value \$$( t_{0.005} \approx 2.68 )$$** \(you can get this from a t-distribution table or calculator).

---

###  **Step 2: Compute the Margin of Error (ME)**
$$
[
ME = t \cdot \frac{s}{\sqrt{n}} = 2.68 \cdot \frac{1.5}{\sqrt{50}} \approx 2.68 \cdot 0.212 \approx 0.568
]
$$
---

### **Step 3: Compute Confidence Interval**
$$
[
\text{CI} = \bar{x} \pm ME = 4 \pm 0.568 = (3.43,\ 4.57)
]
$$

---

###  **Conclusion:**

With 99% confidence, the **true average number of ads** watched by viewers lies between **3.43 and 4.57**.
